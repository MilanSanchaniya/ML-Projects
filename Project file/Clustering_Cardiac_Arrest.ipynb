```python
print(data.info())
```

# 4. Describe dataset in statistic form

```python
data.describe()
```

# 5. Drop under risk column

```python
data.drop("UnderRisk",axis = 1,inplace = True)
```

# 1. Apply Kmean clustering on dataset

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans()
```

```python
model.fit(data)
```

# 2. Print cluster centers

```python
model.cluster_centers_
```

# 3. Create column cluster for predicted labels value

```python
yp = model.fit_predict(data)
```

```python
data["cluster"] = yp
```

# 4. Plot the hierarchical clustering using scipy

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
model1 = linkage(data)
```

```python
dendrogram(model1);
```

# 5. Apply Agglomerative Clustering using number of clusters

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model1 = AgglomerativeClustering()
```

```python
model1.fit(data)
```

# 6. Create label column for predicted cluster label

```python
yp1 = model1.fit_predict(data)
```

```python
data["predicted_cluster"] = yp1
```

# 7. Show label counts

```python
data["predicted_cluster"].value_counts()
```

# 8. Show a silhouette score

```python
from sklearn.metrics import silhouette_samples
```

```python
silhouette_samples(data,yp1)
```
