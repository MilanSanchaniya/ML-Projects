# 1) Perform encoding in columns Gender, Customer_Type, Type_of_Travel, and Class. 

```python
data["Gender"] = data["Gender"].replace(["Male","Female"],[1,0])
```

```python
data["Customer Type"] = data["Customer Type"].replace(['Loyal Customer', 'disloyal Customer'],[1,0])
```

```python
data["Type of Travel"] = data["Type of Travel"].replace(['Personal Travel', 'Business travel'],[0,1])
```

```python
data["Class"] = data["Class"].replace(['Eco Plus', 'Business', 'Eco'],[1,2,0])
```

# 2) Drop the column id and unnamed:_0.1

```python
data.drop("Unnamed: 0.1", axis =1, inplace = True)
```

# 3) Create the features and target Data 

```python
x = data.drop("satisfaction",axis = 1)
y = data["satisfaction"]
```

# 4) Perform scaling on features data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x["Flight Distance"] = scaler.fit_transform(x[["Flight Distance"]])
```

```python
x["Age"] = scaler.fit_transform(x[["Age"]])
```

```python
x["Departure Delay in Minutes"] = scaler.fit_transform(x[["Departure Delay in Minutes"]])
```

```python
x["Arrival Delay in Minutes"] = scaler.fit_transform(x[["Arrival Delay in Minutes"]])
```

# 5) Split the data in training and testing sets 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# 6) Fit the decision tree model with various parameters 

```python
from sklearn.tree import DecisionTreeClassifier,export_text,plot_tree
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

# 7) Create a function to display precision score, recall score, accuracy, classification report, confusion matrix, F1 Score.

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score,classification_report,confusion_matrix,f1_score
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
yp=model.predict(x_test)
```

```python
precision_score(y_test,yp)
```

```python
recall_score(y_test,yp)
```

```python
c = confusion_matrix(y_test,yp)
```

```python
import seaborn as sns
```

```python
sns.heatmap(c,annot=True)
```

```python
print(export_text(model))
```

```python
plot_tree(model,feature_names=x.columns,class_names=["0","1"],filled=True);
```

```python
f1_score(y_test,yp)
```

```python
print(classification_report(y_test,yp))
```
