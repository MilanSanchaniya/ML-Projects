```python
data.dropna(inplace=True,ignore_index=True)
```

```python
data.drop_duplicates(inplace=True,ignore_index=True)
```

```python
data["Date_of_Journey"]=pd.to_datetime(data["Date_of_Journey"])
```

```python
data["Day"]=data["Date_of_Journey"].dt.day
```

```python
data["month"]=data["Date_of_Journey"].dt.month
```

```python
data["year"]=data["Date_of_Journey"].dt.year
```

```python
data.drop("Date_of_Journey",axis =1,inplace = True)
```

```python
data.drop("year",axis =1,inplace = True)
```

```python
data["dep_hour"]=pd.to_datetime(data["Dep_Time"]).dt.hour
```

```python
data["dep_min"]=pd.to_datetime(data["Dep_Time"]).dt.minute
```

```python
data.drop("Dep_Time",axis=1,inplace =True)
```

```python
data["Arrival_hour"]=pd.to_datetime(data["Arrival_Time"]).dt.hour
```

```python
data["Arrival_min"]=pd.to_datetime(data["Arrival_Time"]).dt.minute
```

```python
data.drop("Arrival_Time",axis=1,inplace=True)
```

```python
data.head()
```

```python
data["Duration"].str.split().str[1]
```

```python
data["Duration_hour"]=data["Duration"].str.split().str[0]
```

```python
data["Duration_min"]=data["Duration"].str.split().str[1]
```

```python
data[["Duration_min","Duration_hour"]]=data[["Duration_min","Duration_hour"]].fillna(0)
```

```python
data.head()
```

```python
data[["Duration_hour","Duration_min"]]=data[["Duration_hour","Duration_min"]].replace("[h-m]","",regex=True)
```

```python
data[["Duration_hour","Duration_min"]]=data[["Duration_hour","Duration_min"]].astype(int)
```

```python
data.drop("Duration",axis=1,inplace=True)
```

```python
data.head()
```

```python
data.head()
```

```python
data.groupby("Airline")["Price"].mean()
```

```python
Airline = pd.get_dummies(data["Airline"])
```

```python
Source=pd.get_dummies(data["Source"])
```

```python
Destination =pd.get_dummies(data["Destination"])
```

```python
data["Total_Stops"].unique()
```

```python
data["Total_Stops"]=data["Total_Stops"].replace(['non-stop', '2 stops', '1 stop', '3 stops', '4 stops'],[0,2,1,3,4])
```

```python
data.drop(["Airline","Source","Destination"],axis=True,inplace=True)
```

```python
data=pd.concat([data,Airline,Source,Destination],axis=1)
```

# Multiple linear Regression

```python
x = data.drop("Price",axis=1)
y=data["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
yp = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
r2_score(y_test,yp)
```

```python
root_mean_squared_error(y_test,yp)
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
model1 = Lasso()
```

```python
model1.fit(x_train,y_train)
```

```python
y_predict=model1.predict(x_test)
```
